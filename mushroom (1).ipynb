### Individuo le variabili categoriche e le trasformo in valori con il label encoder

```python
s=(mushroom_data.dtypes == 'object')
object_cols = list(s[s].index)

print (object_cols)
```

```python

label_mushroom_data=mushroom_data.copy()
label_encoder=LabelEncoder()
for col in object_cols:
    label_mushroom_data[col]=label_encoder.fit_transform(label_mushroom_data[col])
```

```python
label_mushroom_data
```

### Individuo target e feature e splitto i dati 

```python
y=label_mushroom_data.class_edible
X=label_mushroom_data.drop("class_edible", axis=1)
```

```python
y
```

```python
X
```

```python

train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=0)
```

```python
train_X
```

```python
train_y
```

##  Modellazione e allenamento

```python
#Definisco il modello 
mushroom_model=DecisionTreeRegressor(random_state=1)
#Alleno il modello
mushroom_model.fit(train_X,train_y)
```

```python
#validazione e calcolo del Mean absosulte error
val_pred=mushroom_model.predict(val_X)

print(mean_absolute_error(val_y,val_pred))
```

## Accuracy

```python
acc_mush_model=accuracy_score(val_y,val_pred)
acc_mush_model
```
